# Creating Line Charts

#### To avoid get a memory error and/or have your browser crash due to to the number of unique values in the prices column, we can create a subset

## Sampling Data

```python
# Create a list holding True/False values to the test np.random.ran <= 0.7
np.random.seed(4)
dev = np.random.rand(len(df_ords_prods_merged)) <= 0.7
```

#### Split the dataframe into two Samples

```python
# Store 70% of the Sample in the Dataframe "big"
big = df_ords_prods_merged[dev]
```

```python
# Store 30% of the Sample in the Dataframe "small"
small = df_ords_prods_merged[~dev]
```

```python
len(df_ords_prods_merged)
```

```python
len(big)+len(small)
```

```python
# Reducing the 'small' sample to only the columns necessary for the chart, namely prices and orders_day_of_week
df_2 = small[['orders_day_of_week','prices']]
```

```python
# Creating the Line Chart for the 'small' Sample
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
# Reducing the 'big' sample to only the columns necessary for the chart, namely prices and orders_day_of_week
df_3 = big[['orders_day_of_week','prices']]
```

```python
# Creating the Line Chart for the 'big' Sample
line_2 = sns.lineplot(data = df_3, x = 'orders_day_of_week',y = 'prices')
```

### Exporting the Line Charts

```python
# Exporting the Line Chart from the 'small' Sample outside of Jupyter
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line.png'))
```

```python
# Exporting the Line Chart from the 'big' Sample outside of Jupyter
line_2.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_2.png'))
```
