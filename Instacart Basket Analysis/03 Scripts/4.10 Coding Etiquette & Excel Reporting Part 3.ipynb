# 6. Create an appropriate visualization to show the distribution of profiles

### Age Group distributions

```python
df_ords_prods_all['age_group'].value_counts()
```

```python
# Setting variable for age group colors

age_colors = ['darkred', 'red', 'crimson', 'indianred', 'lightcoral', 'salmon']
```

```python
# Bar Chart Age Group distribution

age_group_distribution = df_ords_prods_all['age_group'].value_counts().sort_index().plot.bar(color = age_colors)
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)  
```

```python
# Save Visualization
age_group_distribution.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'age_group_distribution.png'))
```

```python
# Age Group Vs Orders distribution

age_group_order_counts = df_ords_prods_all.groupby('age_group')['order_id'].nunique()
```

```python
# Defining Color
order_colors = ['powderblue', 'lightskyblue', 'cornflowerblue', 'royalblue', 'mediumblue', 'navy']
```

```python
# Bar Chart Age Group Vs Orders distribution

age_orders_distribution = age_group_order_counts.plot.bar(color = order_colors)
plt.title("Number of Orders by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45) 
```

```python
# Save Visualization
age_orders_distribution.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'age_orders_distribution.png'))
```

```python
# Age Group Orders trends Vs day of week

age_dow = df_ords_prods_all.groupby(['orders_day_of_week', 'age_group'])['order_id'].nunique().reset_index()
```

```python
pivot_data = age_dow.pivot(index='orders_day_of_week', columns='age_group', values='order_id')
```

```python
# Bar Chart Age Group ordering Vs Day of the week distribution

age_dow_bar = pivot_data.plot.bar(color = order_colors)
plt.title('Age Group Ordering by Day of Week') 
plt.ylabel('Number of Orders')  
plt.xlabel('Day of Week')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

```python
# Save Visualization
age_dow_bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'age_dow_bar.png'))
```
