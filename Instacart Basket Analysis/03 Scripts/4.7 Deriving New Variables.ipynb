# Creating a Flag according to product prices in our df_ords_prods_merged dataframe

## If-else Statements

```python
# Filter Conditions 

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

## If-Statements with the loc() Function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

## Aplying the loc() function to the whole data frame

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Check values count
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
result_1 = []

for value_1 in df_ords_prods_merged["orders_day_of_week"]:
  if value_1 == 0 or value_1 == 1:
    result_1.append("Busiest days")
  elif value_1 == 4 or value_1 == 3:
    result_1.append("Slowest days")
  else:
    result_1.append("Regularly busy")
```

## Question 4
### When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Check values count
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

#### The client wants these grouped into three different segments, we can group the 24 hours of the day as follows: Most orders = 10, 11, 14, 15, 13, 12, 16, 9; Average orders = 17, 8, 18, 19, 20, 7, 21, 22; Fewest orders = 23, 6, 0, 1, 5, 2, 4, 3

```python
# Create a for-loop if statement using the above breakdown
hour_of_day = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    hour_of_day.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    hour_of_day.append("Fewest orders")
  else:
    hour_of_day.append("Average orders")
```

```python
# Print Results
hour_of_day
```

```python
# Creating a new column within the dataframe with the result
df_ords_prods_merged['busiest_period_of_day'] = hour_of_day
```

```python
# Check values count (Frecuency)
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```

#### These frequencies match the sums of the above "ungrouped" frequencies, confirming the correct application of the labels
